# Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
jobs data.
This task will be done in following steps:
1. First get the webpage https://www.naukri.com/
2. Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the
location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.
Note: All of the above steps have to be done in code. No step is to be done manually.

```python
#lets import necessary libraries
import selenium
from selenium import webdriver
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
#calling the driver
driver=webdriver.Chrome(r'C:\Users\Jittu\Desktop\chromedriver.exe')
```

```python
#fetching url
url = "https://www.naukri.com/"
driver.get(url)
```

```python
## finding web element for search job bar using class name , and search loc using absolute xpath and sending keys
search_job = driver.find_element_by_class_name("suggestor-input")
search_loc = driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[3]/div/div/div/input')
search_job.send_keys("Data Analyst")
search_loc.send_keys("Bangalore")
```

```python
## finding web element for search bar using class name , and clicking the search button
search_btn = driver.find_element_by_class_name("qsbSubmit")
search_btn.click()
```

### Extracting required data:

```python
#Extracting all the web-elements
jobtitle_tags = driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')
jobloc_tags = driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi location"]')
company_tags = driver.find_elements_by_xpath('//a[@class="subTitle ellipsis fleft"]')
exp_tags = driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')
```

```python
#checking the length matching of web-elements
print(len(jobtitle_tags),len(jobloc_tags),len(company_tags),len(exp_tags))
```

```python
#Creating empty lists and storing the required data in the list

jobtitle = []
location = []
company = []
experience = []

for i in jobtitle_tags:
    jobtitle.append(i.text)

for i in jobloc_tags:
    location.append(i.text)
    
for i in company_tags:
    company.append(i.text)    
    
for i in exp_tags:
    experience.append(i.text)     
```

```python
#Checking the lengths of appended lists
print(len(jobtitle),len(location),len(company),len(experience))
```

```python
#printing DataFrame
Jobs_df = pd.DataFrame({'Job_title':jobtitle, 'Job_Location':location, 'Company_name':company, 'Experience_required':experience})
Jobs_df
```

```python
Jobs_df.head(10)
```

```python
#Closing the driver
driver.close()
```
